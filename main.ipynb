```python
from dotenv import load_dotenv
import os
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
```

```python
groq_key = os.getenv("GROQ_API_KEY")
```

```python
llm = ChatGroq(
  model="llama-3.3-70b-versatile", 
  temperature=0.1
)
```

```python
parser = StrOutputParser()
```

```python
def chat():
  chat_history = [
    ("system", "You are a helpful assistant. Always answer in a concise manner. Be accurate. Never assume anything."),
  ]

  print("Welcome to the LANGCHAIN GROQ LLM CHATBOT! Type 'exit' to quit.\n")

  while True:
    user_input = input("User: ").strip()

    if user_input.lower() == 'exit':
      print("\nExiting the chatbot. Goodbye!\n")
      break

    #add user message
    chat_history.append(("user", user_input))
    print(f"You: {user_input}")

    #build prompt dynamically
    prompt = ChatPromptTemplate.from_messages(chat_history)

    # prepping the chain of command - send the prompt as input for llm and its output
    # as input for the parser
    chain = prompt | llm | parser

    # get response for user query
    response = chain.invoke({})

    # show and save the assistant's response
    print(f"Bot: {response}\n")

    chat_history.append(("assistant", response))

    print("-"*80)
```

```python
chat()
```
